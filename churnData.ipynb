```python
import matplotlib.pyplot as plt
total_counts = df.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).unstack() * 100
#plot #adjust figsize for better visualization
fig, ax = plt.subplots(figsize=(6, 6))
#plot the bars
total_counts.plot(kind ='bar' ,stacked=True , ax=ax , color = ['#1f77b4','#ff7f0e']) #customize colors if desired
#add percentage labels to the bar 
for p in ax.patches:
    width , height = p.get_width(),p.get_height()
    x , y = p.get_xy()
    ax.text(x + width / 2 , y + height / 2 , f' {height:.1f}%', ha= 'center', va='center')
plt.title('Churn by Senior Citizen (Stacked Bar Chart)')
plt.xlabel('SeniorCitizen')
plt.ylabel('Percenatage(%)')
plt.xticks(rotation=0)
plt.legend(title='Churn',bbox_to_anchor = (1.2,1.2 )) #Customize Legend Location
plt.show()
```

```python
df.columns.values
```

```python
columns = ['PhoneService','MultipleLines','InternetService', 'OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies']
# number of columns for the subplot grid 
n_cols = 3
n_rows = (len(columns) + n_cols - 1 )//n_cols
# create subplots # adjust figsize as needed
fig,axes = plt.subplots(n_rows, n_cols,figsize = (15,n_rows * 4))      
# Flatten the axes array for easy iteration 
axes = axes.flatten()
# Iterate over columns and plot count plots
for i,col in enumerate (columns):
    sns.countplot(x=col, data=df, ax=axes[i], hue="Churn") 
    axes[i].set_title( f' Count Plot of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

for j in range(i+1,len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()

```

#The visualizations show customer churn trends across various telecom services, revealing that fiber optic internet users and those without tech support, online security, or backup services tend to churn more. While most customers have phone service, those with multiple lines show slightly higher churn. Streaming services (TV & movies) do not significantly impact churn. Overall, lack of support and security features strongly correlate with higher customer churn, highlighting their importance in retention strategies.
